## 因子

```python
# 读取月度因子数据
df = pd.read_csv('data/preprocessed/factors_monthly.csv')
df.dropna(inplace=True) 
df.drop(columns='date', inplace=True)
df['year_month'] = pd.to_datetime(df['year_month'], format='%Y-%m')
df.set_index('year_month', inplace=True)
```

```python
print("="*30 + "数据形状" + "="*30 + "\n", df.shape, "\n", "="*64, '\n\n')
print("="*30 + "数据列名" + "="*30 + "\n", df.columns,"\n" , "="*64, '\n\n')
```

### 市值 账面市值比

```python
# 市值和账面市值比
# code=3
BM_Size = pd.read_csv('data/preprocessed/BM_Size_monthly.csv')
BM_Size['year_month'] = pd.to_datetime(BM_Size['year_month'])
BM_Size.set_index('year_month', inplace=True)
BM_Size[BM_Size['code']==3]['Size'].plot()
```
